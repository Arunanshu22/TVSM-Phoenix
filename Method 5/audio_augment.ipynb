```python
# complete audio augment code - 
import librosa
import soundfile as sf
import numpy as np
import pygame

def save_audio(audio, output_path, sr=44100):
    sf.write(output_path, audio, sr)

def load_audio(file_path):
    y, sr = librosa.load(file_path, sr=None)
    return y, sr

def time_stretch(audio, rate):
    return librosa.effects.time_stretch(audio, rate=rate)

def pitch_shift(audio, n_steps):
    return librosa.effects.pitch_shift(audio, sr=44100, n_steps=n_steps)

def speed_change(audio, speed_factor):
    return librosa.effects.preemphasis(audio, coef=speed_factor)

def add_noise(audio, noise_level=0.005):
    noise = np.random.normal(0, noise_level, len(audio))
    return audio + noise

def time_pitch_augmentation(audio, time_stretch_rate, pitch_shift_steps):
    stretched_audio = time_stretch(audio, rate=time_stretch_rate)
    pitched_audio = pitch_shift(stretched_audio, n_steps=pitch_shift_steps)
    return pitched_audio

def random_crop(audio, crop_length):
    if len(audio) <= crop_length:
        return audio
    else:
        start = np.random.randint(0, len(audio) - crop_length)
        return audio[start:start + crop_length]

def dynamic_range_compression(audio, factor=0.5):
    return librosa.effects.preemphasis(audio, coef=factor)

def play_audio(audio_path, sr):
    pygame.mixer.init(frequency=sr)
    pygame.mixer.music.load(audio_path)
    pygame.mixer.music.play()
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

# Specify your input audio file path
input_file_path = 'D:/Outlook Codes/Method 5/audio_files/Arun10Pronounciation.wav'

# Load the original audio
y, sr = load_audio(input_file_path)

# Time stretching (change the rate as needed)
stretched_audio = time_stretch(y, rate=1.2)
save_audio(stretched_audio, "augmented_audio/output_stretched.wav")

# Pitch shifting (change the n_steps as needed)
pitch_shifted_audio = pitch_shift(y, n_steps=2)
save_audio(pitch_shifted_audio, "augmented_audio/output_pitch_shifted.wav")

# Speed change (change the speed factor as needed)
speed_changed_audio = speed_change(y, speed_factor=1.5)
save_audio(speed_changed_audio, "augmented_audio/output_speed_changed.wav")

# Add noise (change the noise level as needed)
noisy_audio = add_noise(y, noise_level=0.01)
save_audio(noisy_audio, "augmented_audio/output_noisy.wav")

# Time and pitch augmentation (change the parameters as needed)
augmented_audio = time_pitch_augmentation(y, time_stretch_rate=1.2, pitch_shift_steps=2)
save_audio(augmented_audio, "augmented_audio/output_augmented.wav")

# Random crop (change the crop length as needed)
crop_length = int(sr * 2)  # 2 seconds
cropped_audio = random_crop(y, crop_length)
save_audio(cropped_audio, "augmented_audio/output_cropped.wav")

# Dynamic range compression (change the compression factor as needed)
compressed_audio = dynamic_range_compression(y, factor=0.3)
save_audio(compressed_audio, "augmented_audio/output_compressed.wav")

# Play the augmented audio samples
play_audio("augmented_audio/output_stretched.wav", sr)
play_audio("augmented_audio/output_pitch_shifted.wav", sr)
play_audio("augmented_audio/output_speed_changed.wav", sr)
play_audio("augmented_audio/output_noisy.wav", sr)
play_audio("augmented_audio/output_augmented.wav", sr)
play_audio("augmented_audio/output_cropped.wav", sr)
play_audio("augmented_audio/output_compressed.wav", sr)
```
