```python
import os
import librosa

def get_audio_duration(file_path):
    try:
        # Load the audio file
        audio, _ = librosa.load(file_path, sr=None)
        
        # Calculate the duration in seconds
        duration = librosa.get_duration(y=audio)
        
        return duration
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

def get_all_audio_durations(folder_path):
    audio_durations = {}
    
    # List all files in the folder
    files = os.listdir(folder_path)

    # Traverse through each file
    for file_name in files:
        if file_name.lower().endswith((".wav", ".mp3")):
            # Construct the full file path
            file_path = os.path.join(folder_path, file_name)
            
            # Get the duration of the audio file
            duration = get_audio_duration(file_path)
            
            # Store the duration in the dictionary
            audio_durations[file_name] = duration

    return audio_durations

# Specify the folder path containing audio files
folder_path = "augmented_audio/"

# Get durations for all audio files in the folder
durations = get_all_audio_durations(folder_path)

# Display the durations
for file_name, duration in durations.items():
    print(f"{file_name}: {duration:.2f} seconds")
```
