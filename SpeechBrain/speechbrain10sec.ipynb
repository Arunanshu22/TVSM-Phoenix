```python
from speechbrain.inference.speaker import SpeakerRecognition
verification = SpeakerRecognition.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb", savedir="pretrained_models/spkrec-ecapa-voxceleb")
```

for 4 sec


```python
for i in range(19):
  for j in range(3):

    score, prediction = verification.verify_files(f"/content/sample_data/sunamdha/sunamdha{i}.wav", f"/content/sample_data/ramya/ramya{j}.wav") # Same Speaker
    print (f"Sunamdha{i}  and ramya{j} ",score,prediction)
```
