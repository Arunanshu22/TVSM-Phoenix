```python
import os
import pyaudio
import sounddevice as sd
import wave
import numpy as np
import time
import scipy.io.wavfile as wav
from speechbrain.inference.speaker import SpeakerRecognition

# Function to record audio and save it as a .wav file
def record_audio(file_path, duration=10):
    CHUNK = 1024
    FORMAT = pyaudio.paInt16
    CHANNELS = 2
    RATE = 48000
    RECORD_SECONDS = duration
    WAVE_OUTPUT_FILENAME = file_path
    audio = pyaudio.PyAudio()
    stream = audio.open(format=FORMAT, channels=CHANNELS,
                        rate=RATE, input=True,
                        frames_per_buffer=CHUNK)
    print("Recording...")
    frames = []
    for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(data)
    print("Finished recording.")
    stream.stop_stream()
    stream.close()
    audio.terminate()
    wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
    wf.setnchannels(CHANNELS)
    wf.setsampwidth(audio.get_sample_size(FORMAT))
    wf.setframerate(RATE)
    wf.writeframes(b''.join(frames))
    wf.close()

def show_enrolled_users():
    print("List of enrolled users:")
    with open("users.txt", "r") as f:
        for line in f:
            # Splitting each line into username and filepath
            username, filepath = line.strip().split(": ")
            print(f"{username}")

def enroll_new_user():
    while True:
        user_name = input("Enter user name: ")
        if os.path.exists("users.txt"):
            with open("users.txt", "r") as f:
                if user_name in f.read():
                    print("User already exists. Please choose a different username.")
                    continue

        file_name = f"{user_name}.wav"
        predict_name = f"{user_name}_predict.wav"
        file_path = os.path.join("final_standard", file_name)
        predict_path = os.path.join("predictTemp", predict_name)

        print("Please get ready to speak. Recording will start in 5 seconds...")
        for i in range(5, 0, -1):
            print(f"Recording starts in {i} seconds...")
            time.sleep(1)
        print("Recording... Speak for 10 seconds. This will be your standard voice to be compared later on...")
        record_audio(file_path)

        print("Please get ready to speak. Recording will start in 5 seconds...")
        for i in range(5, 0, -1):
            print(f"Recording starts in {i} seconds...")
            time.sleep(1)
        print("Recording... Speak for 10 seconds. This is for the validation of previous files authenticity")
        record_audio(predict_path)

        verification = SpeakerRecognition.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb", savedir="pretrained_models/spkrec-ecapa-voxceleb")
        score, prediction = verification.verify_files(file_path, predict_path)
        if score < 0.6:
            print(f"Verification score ({score}) is less than 0.6. Re-enrolling...")
            if os.path.exists(file_path):
                os.remove(file_path)
            if os.path.exists(predict_path):
                os.remove(predict_path)
        else:
            print(f"User '{user_name}' enrolled successfully with verification score: {score}")
            # Write user information to the users.txt file
            with open("users.txt", "a") as f:
                f.write(f"{user_name}: {file_path}\n")
            break

def validate_user():
    unknown_file = "unknown.wav"
    unknown_file_path = os.path.join("unknownVoices", unknown_file)
    print("Please get ready to speak. Recording will start in 5 seconds...")
    for i in range(5, 0, -1):
        print(f"Recording starts in {i} seconds...")
        time.sleep(1)
    print("Recording... Speak for 10 seconds. This will be your standard voice to be compared later on...")
    record_audio(unknown_file_path)

    found_user = False
    with open("users.txt", "r") as f:
        for line in f:
            username, filepath = line.strip().split(": ")
            verification = SpeakerRecognition.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb", savedir="pretrained_models/spkrec-ecapa-voxceleb")
            score, prediction = verification.verify_files(filepath, unknown_file_path)
            print(score)
            if score > 0.6:
                print(f"User found: {username}")
                found_user = True
                break
    if not found_user:
        print("Unknown user")

def change_standard_voice():
    print("Changing standard voice...")
    # Functionality to change standard voice goes here

def main():
    while True:
        print("\nTO DO - ")
        print("1) Press 1 to see list of enrolled users")
        print("2) Press 2 to enroll a new user")
        print("3) Press 3 to validate a user")
        print("4) Press 4 to change standard voice")
        print("5) Press 5 to quit")

        choice = input("Enter your choice: ")

        if choice == '1':
            show_enrolled_users()
        elif choice == '2':
            enroll_new_user()
        elif choice == '3':
            validate_user()
        elif choice == '4':
            change_standard_voice()
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```
